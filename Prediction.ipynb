```python
import pandas as pd
import numpy as np
import math
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('./train.csv')
df
```

# Pre-processing

```python
df.isnull().sum()
```

```python
## Delete Un-necessary Column
df.drop(['PassengerId','Name','Ticket','Cabin'],axis = 1 ,inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['Age'].isnull() == True]
```

```python
df['Age'].fillna(np.mean(df['Age']),inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['Embarked'].isnull()]
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace = True)
```

```python
df[df['Embarked'].isnull()]
```

```python
df.isnull().sum()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'] = np.where(df['Sex'] == 'male',0,1)
df['Sex'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
df['Age'] = df['Age'].apply(lambda x : math.ceil(x))
df['Age'].unique()
```

```python
df['Parch'].unique()
```

```python
df['Fare'] = df['Fare'].apply(lambda x: round(x,2)) 
```

```python
df['Fare'].unique()
```

```python
# Encoding Embarked Column
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Embarked']]).toarray()
encoded_df = pd.DataFrame(encoded,columns = encoder.get_feature_names_out())
df = pd.concat([df,encoded_df],axis = 1)
df.drop('Embarked',axis = 1,inplace = True)
```

```python
df
```

#### Divide Independent and Dependent Variable

```python
X = df.drop('Survived',axis = 1)
y = df['Survived']
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 40,test_size = 0.2)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
import warnings
warnings.filterwarnings('ignore')
```

#### intializing Cross Validation

```python
cv = KFold(n_splits= 5)
cv
```

#### HyperParameter Tunnig

```python
parameter = {'penalty':['l1','l2','elasticnet'],'C' : [1,5,10,15,20]}
clf = GridSearchCV(LogisticRegression(),param_grid = parameter,cv = cv)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
## It is the mean of score of cross validation
clf.best_score_
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
